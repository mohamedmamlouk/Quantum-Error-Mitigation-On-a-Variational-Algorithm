```python
from qiskit.test.mock import FakeLima
import datetime
from dateutil.tz import tzoffset

# Create an instance of the FakeLima backend
backend = FakeLima()

# Fetch the backend properties
properties = backend.properties()

# Function to extract error information
def extract_error_info(properties):
    error_info = {}
    
    # Readout errors for each qubit
    for qubit in properties.qubits:
        for item in qubit:
            if item.name == 'readout_error':
                error_info[f"Qubit {properties.qubits.index(qubit)} Readout Error"] = item.value

    # Gate errors for each gate on each qubit
    for gate in properties.gates:
        gate_name = gate.name
        for param in gate.parameters:
            if param.name == 'gate_error':
                error_info[f"{gate_name} Gate Error"] = param.value
    
    return error_info

# Extract and print error information
error_info = extract_error_info(properties)
for key, value in error_info.items():
    print(f"{key}: {value}")
```

```python
from qiskit.providers.aer.noise import NoiseModel
from qiskit.test.mock import FakeLima

# Get the backend
backend = FakeLima()

# Generate a noise model from the backend
noise_model = NoiseModel.from_backend(backend)

# Printing all quantum errors in the noise model
errors = noise_model.to_dict()['errors']

print("Quantum Errors in the Noise Model:")
for error in errors:
    print(f"Error on gates: {error['operations']}")
    print(f"Error on qubits: {error['gate_qubits']}")
    print(f"Error specifics: {error['probabilities']}, {error['instructions']}")
    print("-----------")
```
